```python
import pandas as pd
```

```python
df = pd.read_excel("./Defunciones/2020_excel.xlsx")
df.info()
```

```python
df = df.iloc[6:,:]
df.head()
```

```python
new_columns = df.iloc[0]
print(new_columns)
df.columns= new_columns
df = df.iloc[1:,:]
df.reset_index(inplace=True)
df
```

```python
df.head()
```

```python
df= df.iloc[:,1:]
df.head()
```

```python
df_interes_1 = df.iloc[0:1 , :]
df_interes_2 = df.iloc[2:4 , :]
df_interes_3 = df.iloc[71:78,:]


print(type(df_interes_1))
print(type(df_interes_2))
print(type(df_interes_3))
```

```python
enfermedades_respiratorias = pd.concat([df_interes_1 , df_interes_2 , df_interes_3] , ignore_index=True)
display(enfermedades_respiratorias)
```

```python
enfermedades_respiratorias_T = enfermedades_respiratorias.T
display(enfermedades_respiratorias_T)
```

```python
enfermedades_respiratorias.to_csv("./Defunciones/Enfermedades_respiratorias_vs_meses.csv" , sep= ';' ,index= False)
```

```python
enfermedades_respiratorias_T.to_csv("./Defunciones/Meses_vs_enfermedades_respiratorias.csv" , sep= ';' , index= True)
```
