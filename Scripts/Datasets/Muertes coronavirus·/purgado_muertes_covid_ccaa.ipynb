### Modificar fecha

```python
def modificacion_fecha(df):
    fecha_mes = df['Periodo']
    año = df['Año']
    lista_fechas = []
    for pos , fecha_mensual in enumerate(fecha_mes):
        #print(fecha_mensual)
        lista = fecha_mensual.split(' ')
        #print(lista)
        if len(lista) == 4:
            dia = lista[1]
            mes_letra = lista[3]
            if mes_letra == 'enero':
                mes = '01'
            elif mes_letra == 'febrero':
                mes = '02'
            elif mes_letra == 'marzo':
                mes = '03'
            elif mes_letra == 'abril':
                mes = '04'
            elif mes_letra == 'mayo':
                mes = '05'
            elif mes_letra == 'junio':
                mes = '06'
            elif mes_letra == 'julio':
                mes = '07'
            elif mes_letra == 'agosto':
                mes = '08'
            elif mes_letra == 'septiembre':
                mes = '09'
            elif mes_letra == 'octubre':
                mes = '10'
            elif mes_letra == 'noviembre':
                mes = '11'
            elif mes_letra == 'diciembre':
                mes = '12'

            fecha_lista = [año[pos] , mes , dia]


        else:
            fecha_lista = ['2021' , '01' , '01']

        fecha = '/'.join(fecha_lista)
        lista_fechas.append(fecha)
    

    df['Fecha'] = lista_fechas

    return df
```

```python
lista_ccaa_final = []
contador = 0
for ccaa in lista_ccaa:
    ccaa = ccaa.dropna()
    ccaa = modificacion_fecha(ccaa)
    lista_ccaa_final.append(ccaa.iloc[: , 2:])
    contador += 1
```

### Agrupacion en semanas

```python
muertes_coronavirus_dia = muertes_coronavirus_dia.fillna(0)
```

```python
muertes_coronavirus_dia.tail()
```

```python

muertes_coronavirus_dia = muertes_coronavirus_dia.loc[: , 'andalucia':] #Eliminar columna 'Fecha'
    
ultimo_lunes = 725
muertes_coronavirus_dia = muertes_coronavirus_dia.iloc[:ultimo_lunes , :] #Eliminar dias de semana incompleta (en última posicion)
muertes_coronavirus_semana_ccaa = pd.DataFrame()

for ccaa in lista_ccaa_nombre:
        
    semanas = []
    muertes_coronavirus_semanas = []
    semana = 1
    posicion_lunes = 4
        
    while posicion_lunes < ultimo_lunes:
        muertes_coronavirus_semana = muertes_coronavirus_dia.iloc[posicion_lunes].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes + 1].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes+2].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes+3].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes+4].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes+5].loc[ccaa] + muertes_coronavirus_dia.iloc[posicion_lunes + 6].loc[ccaa]
        muertes_coronavirus_semanas.append(muertes_coronavirus_semana)
        semanas.append(semana)
        semana += 1
        posicion_lunes += 7
        
        
    muertes_coronavirus_semana_ccaa['Semana'] = semanas
    muertes_coronavirus_semana_ccaa[ccaa] = muertes_coronavirus_semanas

muertes_coronavirus_semana_ccaa
           
```
