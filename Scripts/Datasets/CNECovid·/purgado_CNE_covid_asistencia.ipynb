### Reorganizacion formato de fecha

```python
new_fechas = []
for fecha in asistencia['Fecha']:
    fecha_div = fecha.split('/')
    new_fecha = '/'.join([fecha_div[2] , fecha_div[1] , fecha_div[0]])
    new_fechas.append(new_fecha)
asistencia['Fecha'] = new_fechas
asistencia.head()
```

### Agrupacion por dia (total en España)

```python
asistencia_dia_españa = asistencia.groupby('Fecha')['TOTAL_CAMAS' , 'OCUPADAS_COVID19' , 'OCUPADAS_NO_COVID19' , 'INGRESOS_COVID19' , 'ALTAS_24h_COVID19'].sum()
asistencia_dia_españa = asistencia_dia_españa.reset_index()
asistencia_dia_españa.head()
```

```python
asistencia_dia_españa.tail(20)
```

##### Agrupacion en semanas

```python
def agrupar_semanas(df , new_df):
    
    ultimo_lunes = 513
    asistencia = df.iloc[:ultimo_lunes , :] #Eliminar dias de semana incompleta (en última posicion)
    lista_columnas = ['TOTAL_CAMAS' , 'OCUPADAS_COVID19' , 'OCUPADAS_NO_COVID19' , 'INGRESOS_COVID19' , 'ALTAS_24h_COVID19']

    for columna in lista_columnas:
            
        semanas = []
        asistencia_semanas = []
        semana = 32
        posicion_lunes = 2
            
        while posicion_lunes < ultimo_lunes:
            asistencia_media_semana = round((df.iloc[posicion_lunes].loc[columna] + df.iloc[posicion_lunes + 1].loc[columna] + df.iloc[posicion_lunes+2].loc[columna] + df.iloc[posicion_lunes+3].loc[columna] + df.iloc[posicion_lunes+4].loc[columna] + df.iloc[posicion_lunes+5].loc[columna] + df.iloc[posicion_lunes + 6].loc[columna])/7 , 0)
            asistencia_semanas.append(int(asistencia_media_semana))
            semanas.append(semana)
            semana += 1
            posicion_lunes += 7
            
            
        new_df['Semana'] = semanas
        new_df[columna] = asistencia_semanas

    return new_df
```

```python
# Medias semanales en España del total de camas, ocupadas por pacientes covid, pacientes no covid, ingresos por covid y altas por covid
asistencia_semana_españa = pd.DataFrame()
agrupar_semanas(asistencia_dia_españa , asistencia_semana_españa)
```

### Agrupacion por ccaa

##### Agrupacion de fechas (equivalente a sumar los datos de las provincias)

```python
ccaa_dfs_dia = []
for ccaa_df in ccaa_dfs:
    ccaa_df_dia = ccaa_df.groupby('Fecha')['TOTAL_CAMAS' , 'OCUPADAS_COVID19' , 'OCUPADAS_NO_COVID19' , 'INGRESOS_COVID19' , 'ALTAS_24h_COVID19'].sum()
    ccaa_df_dia = ccaa_df_dia.reset_index()
    ccaa_dfs_dia.append(ccaa_df_dia)
```

```python
ccaa_dfs_dia[0]
```

##### Transofrmacion de dfs (generar 5 dfs con las columnas de datos)

```python
def transformacion_df_ccaa(df , columna):
    for pos , ccaa_df in enumerate(ccaa_dfs_dia):
        nombre = ccaa_name[pos]
        df['Fecha'] =  ccaa_df['Fecha']
        df[nombre] = ccaa_df[columna]

    return df
```

### Agrupacion en semanas

```python

def agrupar_semanas(df , new_df):
    
    ultimo_lunes = 513
    asistencia = df.iloc[:ultimo_lunes , :] #Eliminar dias de semana incompleta (en última posicion)
    lista_columnas = list(df.columns.values) #Toma como columnas las ya presentes en las dfs
    lista_columnas = lista_columnas[1:] #Menos la columna inicial de 'Fecha', que se cambia a 'Semana'

    for columna in lista_columnas:
            
        semanas = []
        asistencia_semanas = []
        semana = 32
        posicion_lunes = 2
            
        while posicion_lunes < ultimo_lunes:
            asistencia_media_semana = round((df.iloc[posicion_lunes].loc[columna] + df.iloc[posicion_lunes + 1].loc[columna] + df.iloc[posicion_lunes+2].loc[columna] + df.iloc[posicion_lunes+3].loc[columna] + df.iloc[posicion_lunes+4].loc[columna] + df.iloc[posicion_lunes+5].loc[columna] + df.iloc[posicion_lunes + 6].loc[columna])/7 , 0)
            asistencia_semanas.append(int(asistencia_media_semana))
            semanas.append(semana)
            semana += 1
            posicion_lunes += 7
            
            
        new_df['Semana'] = semanas
        new_df[columna] = asistencia_semanas

    return new_df
```
