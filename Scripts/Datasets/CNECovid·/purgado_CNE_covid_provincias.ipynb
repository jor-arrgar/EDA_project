### Agrupacion en semanas

```python
# Seleccion de fechas adecuadas (eliminacion de la cola)

def eliminar_cola(casos):
    casos = casos.reset_index()
    casos = casos.iloc[:726 , :]
    return casos.tail()
    
```

```python
# Funcion de agrupacion
def agrupar_semanas(provincia , nombre):
    casos_semanales = []
    posicion_lunes = 5
    semana = 2
    semanas = []

    while len(semanas) <= 102:
        casos_semana = provincia.iloc[posicion_lunes].loc['num_casos'] + provincia.iloc[posicion_lunes + 1].loc['num_casos'] + provincia.iloc[posicion_lunes+2].loc['num_casos'] + provincia.iloc[posicion_lunes+3].loc['num_casos'] + provincia.iloc[posicion_lunes+4].loc['num_casos'] + provincia.iloc[posicion_lunes+5].loc['num_casos'] + provincia.iloc[posicion_lunes + 6].loc['num_casos'] 
        casos_semanales.append(casos_semana)
        semanas.append(semana)

        posicion_lunes += 7
        semana += 1
        casos_semana = 0

    casos_semanales= pd.Series(casos_semanales)
    semanas = pd.Series(semanas)

    #covid_semanal = pd.DataFrame()
    #covid_semanal['Semana'] = semanas
    covid_semanal[nombre] = casos_semanales
    return covid_semanal.head()
```
