### Agrupación en semanas

```python
casos_ccaa.info()
```

Todos las comunidades presentan datos para las mismas fechas

```python
def agrupar_semanas(dataframe , ccaa):
    casos_semanales = []
    posicion_lunes = 5
    semana = 2
    semanas = []

    while len(semanas) <= 102:
        casos_semana = dataframe.iloc[posicion_lunes].loc['num_casos'] + dataframe.iloc[posicion_lunes + 1].loc['num_casos'] + dataframe.iloc[posicion_lunes+2].loc['num_casos'] + dataframe.iloc[posicion_lunes+3].loc['num_casos'] + dataframe.iloc[posicion_lunes+4].loc['num_casos'] + dataframe.iloc[posicion_lunes+5].loc['num_casos'] + dataframe.iloc[posicion_lunes + 6].loc['num_casos']
        casos_semanales.append(casos_semana)
        semanas.append(semana)

        posicion_lunes += 7
        semana += 1
        casos_semana = 0

    casos_semanales= pd.Series(casos_semanales)
    semanas = pd.Series(semanas)


    covid_semanal['Semana'] = semanas
    covid_semanal[ccaa] = casos_semanales
```

```python
lista_dataframes = [casos_andalucia , casos_aragon , casos_asturias , casos_baleares , casos_canarias , casos_cantabria , casos_cataluña , casos_clm , casos_cyl , casos_ceuta , casos_valencia, casos_extrem , casos_galicia ,casos_rioja, casos_madrid , casos_melilla , casos_murcia , casos_navarra , casos_pais_vasco]
lista_ccaa = ['Andalucia' , 'Aragón' , 'Asturias' , 'Baleares' , 'Canarias' , 'Cantabria' , 'Cataluña' , 'Castilla-La mancha' , 'Castilla y León' , 'Ceuta' , 'Com. Valenciana' , 'Extremadura' , 'Galicia' , 'La Rioja' , 'Madrid' , 'Melilla' , 'Murcia' , 'Navarra' , 'Pais Vasco']
covid_semanal = pd.DataFrame()
```

```python
posicion = 0
for df in lista_dataframes:
    nombre = lista_ccaa[posicion]
    agrupar_semanas(df , nombre)
    posicion += 1
    
```
