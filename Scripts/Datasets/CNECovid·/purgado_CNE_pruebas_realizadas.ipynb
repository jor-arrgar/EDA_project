##### Modificacion de fecha para ordenar cronológicamente

```python
pruebas_dia.columns.values
```

```python
fechas_num = []
dias = []
meses = []
años = []

for fecha in pruebas_dia['FECHA_PRUEBA']:

    dia = fecha[:2]
    dias.append(int(dia))
    año = int(fecha[5:])
    años.append(año)
    mes_letras = fecha[2:5]

    if mes_letras == 'JAN':
        mes_num = '01'
    elif mes_letras == 'FEB':
        mes_num = '02'
    elif mes_letras == 'MAR':
        mes_num = '03'
    elif mes_letras == 'APR':
        mes_num = '04'
    elif mes_letras == 'MAY':
        mes_num = '05'
    elif mes_letras == 'JUN':
        mes_num = '06'
    elif mes_letras == 'JUL':
        mes_num = '07'
    elif mes_letras == 'AUG':
        mes_num = '08'
    elif mes_letras == 'SEP':
        mes_num = '09'
    elif mes_letras == 'OCT':
        mes_num = '10'
    elif mes_letras == 'NOV':
        mes_num = '11'
    elif mes_letras == 'DEC':
        mes_num = '12'
    meses.append(int(mes_num))

    fecha_num = [str(año) , mes_num , str(dia)]
    fecha_num = '/'.join(fecha_num)
    fechas_num.append(fecha_num)
pruebas_dia = pd.DataFrame()
pruebas_dia['FECHA'] = fechas_num
pruebas_dia = pruebas_dia.sort_values(by= 'FECHA' , ascending= True)

```

```python
def unificar_provincias(pruebas_df , columna):
    
    conteo = 0
    for df_provincia in lista_provincias:
        nombre = lista_provincias_nombre[conteo]
        valores = list(df_provincia[columna])
        if len(df_provincia['FECHA'].values) == 654:
            fechas = list(alava['FECHA'])
            pruebas_df['Fecha'] = fechas
            pruebas_df[nombre] = valores
        elif len(df_provincia['FECHA'].values) == 653:
            fechas = list(df_provincia['FECHA'])
            prov_fecha_diferente = pd.DataFrame()
            prov_fecha_diferente['Fecha'] = fechas
            prov_fecha_diferente[nombre] = valores

            pruebas_df = pd.merge(pruebas_df , prov_fecha_diferente , on= 'Fecha' , how= 'outer')

        conteo += 1
    pruebas_df = pruebas_df.fillna(0)
    pruebas_df = pruebas_df.sort_values(by= 'Fecha' , ascending= True)
    pruebas_df = pruebas_df.loc[: , 'Fecha':]
    pruebas_df = pruebas_df.reset_index()

    return pruebas_df
```

### Agrupacion en semanas

```python
def agrupar_semanas(df , new_df):
    df = df.reset_index()
    df = df.loc[: , 'alava':] #Eliminar columna 'Fecha'
    
    ultimo_lunes = 652
    df = df.iloc[:ultimo_lunes , :] #Eliminar dias de semana incompleta (en última posicion)


    for provincia in lista_provincias_nombre:
        
        semanas = []
        pruebas_semanas = []
        semana = 12
        posicion_lunes = 1
        
        while posicion_lunes < ultimo_lunes:
            pruebas_semana = df.iloc[posicion_lunes].loc[provincia] + df.iloc[posicion_lunes + 1].loc[provincia] + df.iloc[posicion_lunes+2].loc[provincia] + df.iloc[posicion_lunes+3].loc[provincia] + df.iloc[posicion_lunes+4].loc[provincia] + df.iloc[posicion_lunes+5].loc[provincia] + df.iloc[posicion_lunes + 6].loc[provincia]
            pruebas_semanas.append(pruebas_semana)
            semanas.append(semana)
            semana += 1
            posicion_lunes += 7
            

        new_df['Semana'] = semanas
        new_df[provincia] = pruebas_semanas
        
    return df
```
