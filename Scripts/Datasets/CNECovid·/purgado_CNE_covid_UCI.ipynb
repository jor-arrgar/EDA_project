```python
def agrupar_edades_y_semanas(provincia , nombre):
    casos_semanales = []
    hospitalizaciones_semanales = []
    uci_semanales = []
    fallecimientos_semanales = []

    semana = 2
    semanas = []
    posicion_lunes = 5

    casos_dia = provincia.groupby('fecha')['num_casos'].sum()
    casos_dia = pd.DataFrame(casos_dia)
    casos_dia = casos_dia.fillna(0)
    fechas = list(casos_dia.index.values)

    hospitalizaciones_diarias = provincia.groupby('fecha')['num_hosp'].sum()
    hospitalizaciones_diarias = pd.DataFrame(hospitalizaciones_diarias)
    hospitalizaciones_diarias = hospitalizaciones_diarias.fillna(0) 

    ucis_diarias = provincia.groupby('fecha')['num_uci'].sum()
    ucis_diarias = pd.DataFrame(ucis_diarias)
    ucis_diarias = ucis_diarias.fillna(0)

    muertes_diarias = provincia.groupby('fecha')['num_def'].sum()
    muertes_diarias = pd.DataFrame(muertes_diarias)
    muertes_diarias = ucis_diarias.fillna(0)
    
    while len(semanas) <= 102:
        #Numero de casos
        try:
            casos_semana = casos_dia.iloc[posicion_lunes , 0] + casos_dia.iloc[posicion_lunes + 1 , 0] + casos_dia.iloc[posicion_lunes+2 , 0] + casos_dia.iloc[posicion_lunes+3 , 0] + casos_dia.iloc[posicion_lunes+4 , 0] + casos_dia.iloc[posicion_lunes+5 , 0] + casos_dia.iloc[posicion_lunes + 6 , 0]
            casos_semanales.append(casos_semana)
            #print(casos_semanales)
        except:
            pass
        #Numero de hospitalizaciones
        try:
            hospitalizaciones_semana = hospitalizaciones_diarias.iloc[posicion_lunes , 0] + hospitalizaciones_diarias.iloc[posicion_lunes + 1 , 0] + hospitalizaciones_diarias.iloc[posicion_lunes+2 , 0] + hospitalizaciones_diarias.iloc[posicion_lunes+3 , 0] + hospitalizaciones_diarias.iloc[posicion_lunes+4 , 0] + hospitalizaciones_diarias.iloc[posicion_lunes+5 , 0] + hospitalizaciones_diarias.iloc[posicion_lunes + 6 , 0]
            hospitalizaciones_semanales.append(hospitalizaciones_semana)
        except:
            pass
        #Numero de UCIs
        try:
            ucis_semana = ucis_diarias.iloc[posicion_lunes , 0] + ucis_diarias.iloc[posicion_lunes + 1 , 0] + ucis_diarias.iloc[posicion_lunes+2 , 0] + ucis_diarias.iloc[posicion_lunes+3 , 0] + ucis_diarias.iloc[posicion_lunes+4 , 0] + ucis_diarias.iloc[posicion_lunes+5 , 0] + ucis_diarias.iloc[posicion_lunes + 6 , 0]
            uci_semanales.append(ucis_semana)
        except:
            pass
        #Numero de fallecidos
        try:
            muertes_semana = muertes_diarias.iloc[posicion_lunes , 0] + muertes_diarias.iloc[posicion_lunes + 1 , 0] + muertes_diarias.iloc[posicion_lunes+2 , 0] + muertes_diarias.iloc[posicion_lunes+3 , 0] + muertes_diarias.iloc[posicion_lunes+4 , 0] + muertes_diarias.iloc[posicion_lunes+5 , 0] + muertes_diarias.iloc[posicion_lunes + 6 , 0]
            fallecimientos_semanales.append(muertes_semana)
        except:
            pass
        semanas.append(semana)
        semana += 1
        posicion_lunes += 7




    casos_semanales = pd.Series(casos_semanales)
    hospitalizaciones_semanales = pd.Series(hospitalizaciones_semanales)
    uci_semanales = pd.Series(uci_semanales)
    fallecimientos_semanales = pd.Series(fallecimientos_semanales)
    semanas = pd.Series(semanas)


    nombre['Semana'] = semanas
    nombre['Casos'] = casos_semanales
    nombre['Hospitalizaciones'] = hospitalizaciones_semanales
    nombre['Ingresos UCI'] = uci_semanales
    nombre['Fallecimientos'] = fallecimientos_semanales

    return nombre
```
