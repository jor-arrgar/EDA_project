### Funcion de tratamiento de los dataframes

```python
def tratamiento_df(df , ccaa):
    #Eliminar 'Semana'
    semanas = []
    for periodo in df['Periodo']:
        periodo_lista = periodo.split(' ')
        semanas.append(periodo_lista[1])
    df['Semana'] = semanas

    #Transformar tasa en 'float'
    tasa = []
    for valor in df['Gripe']:
        if valor != '0':
            lista_valor = valor.split(',')
            new_valor = float('.'.join(lista_valor))
        else:
            new_valor = 0.0
        tasa.append(new_valor)
    df[ccaa] = tasa

    #Limpiar columnas sobrantes
    tasa_ = df.iloc[:,-2:]
    anyos = df['Año']
    df = pd.concat([anyos , tasa_] , axis= 1)

    #Guardado en csv
    df.to_csv('./purgado/tasa_semanal_gripe_' + ccaa + '.csv' , sep= ';' , index= False)
```

Las comunidades autonomas no presentan el mismo numero de semanas, por eso se tratan cada una por separado
