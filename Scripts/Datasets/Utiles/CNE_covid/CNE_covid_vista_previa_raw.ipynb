```python
#Total de antigenos positivos por semana
antigenos_pos = []
lista_prov_ant = list(antigenos_positivos_prov.columns.values)
lista_prov_ant = lista_prov_ant[1:]

semanas = list(antigenos_positivos_prov['Semana'])

for pos , semana in enumerate(semanas):
    suma = 0
    for provincia in lista_prov_ant:
        suma += antigenos_positivos_prov.iloc[pos].loc[provincia]
    antigenos_pos.append(suma)
len(semanas)
```

```python
antigenos_tot = []
lista_prov = list(antigenos_totales_prov.columns.values)

semanas = list(antigenos_positivos_prov['Semana'])

for pos , semana in enumerate(semanas):
    suma = 0
    for provincia in lista_prov:
        suma = suma + antigenos_totales_prov.iloc[pos].loc[provincia]
    antigenos_tot.append(suma)

```

```python
pcr_pos = []
lista_prov = list(pcr_pos_prov.columns.values)


semanas = list(antigenos_positivos_prov['Semana'])

for pos , semana in enumerate(semanas):
    suma = 0
    for provincia in lista_prov:
        suma += pcr_pos_prov.iloc[pos].loc[provincia]
    pcr_pos.append(suma)
```
