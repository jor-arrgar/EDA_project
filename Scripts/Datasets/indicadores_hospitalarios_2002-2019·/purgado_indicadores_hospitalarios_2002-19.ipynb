```python
import pandas as pd
```

```python
indicadores = pd.read_csv('./Indicadores_hospitalarios_2002-2019.csv' , sep= ';')
indicadores.info()
```

```python
indicadores
```

```python
indicadores = indicadores.iloc[: , :-1]
indicadores.head()
```

```python
num_indicador = set(list(indicadores['indicador']))
nombre_indicador = set(list(indicadores['nombre de indicador']))
año = set(list(indicadores['a?o']))
territorio = list(set(list(indicadores['ccaa'])))
territorio.sort()
sexo = set(list(indicadores['sexo']))

print('- nº indicador: ' , num_indicador)
print('- nombre_indicador: ' , nombre_indicador)
print('- año: ' , año)
print('- territorio: ' , territorio)
print('- sexo: ' , sexo)
```

### Camas por 1000 habitantes

```python
camas = indicadores[indicadores['nombre de indicador'] == 'Camas hospitalarias en funcionamiento por 1000 hab.']
camas.head()
```

```python
dfs_territorios = []

for ccaa in territorio:
    df_territorio = camas[camas['ccaa'] == ccaa]
    dfs_territorios.append(df_territorio)
    
```

```python
for df in dfs_territorios:
    df.info()
```

```python
dfs_territorios[3]
```

```python
for df in dfs_territorios:
    valores = []
    for valor in df['valor']:
        valores.append(round(valor , 2))
    df['valores_2_dec'] = valores
```

```python
dfs_territorios[0]
```

```python
camas_x_1000_hab = pd.DataFrame()

for pos , df in enumerate(dfs_territorios):
    df_res = df.reset_index()
    camas_x_1000_hab['año'] = df_res['a?o']
    camas_x_1000_hab[territorio[pos]] = df_res['valores_2_dec']
camas_x_1000_hab
```

### Gasto sanitario publico por habitante

```python
gasto_publico = indicadores[indicadores['nombre de indicador'] == 'Gasto sanitario p?blico por habitante']
gasto_publico = gasto_publico.reset_index()
gasto_publico
```

```python

valores = []
for valor in gasto_publico['valor']:
    valores.append(round(valor , 2))
print(valores)
gasto_publico['valores_2_dec'] = valores
gasto_publico.head()
```

```python
gasto_publico_españa = pd.concat([gasto_publico.loc[: , 'a?o'] , gasto_publico.loc[: , 'valores_2_dec']] , axis= 1)
gasto_publico_españa
```

### Porcentaje de gasto en atencion primaria

```python
porcentaje_atencion_primaria = indicadores[indicadores['nombre de indicador'] == 'Porcentaje del gasto sanitario en atenci?n primaria']
porcentaje_atencion_primaria
```

```python
dfs_territorios = []

for ccaa in territorio:
    df_territorio = porcentaje_atencion_primaria[porcentaje_atencion_primaria['ccaa'] == ccaa]
    dfs_territorios.append(df_territorio)
```

```python
for df in dfs_territorios:
    df.info()
```

```python
dfs_territorios[0]
```

```python
for df in dfs_territorios:
    valores = []
    for valor in df['valor']:
        valores.append(round(valor , 2))
    df['valores_2_dec'] = valores
```

```python
dfs_territorios[0]
```

```python
gasto_sanitario_atencion_primaria_porcentaje = pd.DataFrame()

for pos , df in enumerate(dfs_territorios):
    df_res = df.reset_index()
    gasto_sanitario_atencion_primaria_porcentaje['año'] = df_res['a?o']
    gasto_sanitario_atencion_primaria_porcentaje[territorio[pos]] = df_res['valores_2_dec']
gasto_sanitario_atencion_primaria_porcentaje
```

### Guardado en csv

```python
camas_x_1000_hab.to_csv('./purgado/camas_x_1000_hab.csv' , sep= ';' , index= False)
gasto_publico_españa.to_csv('./purgado/gasto_publico_sanitario_españa.csv' , sep= ';' , index= False)
gasto_sanitario_atencion_primaria_porcentaje.to_csv('./purgado/%_gasto_publico_atencion_primaria.csv' , sep= ';' , index= False)
```
